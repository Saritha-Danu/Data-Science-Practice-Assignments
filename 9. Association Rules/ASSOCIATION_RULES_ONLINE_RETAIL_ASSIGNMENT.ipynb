# Data Preprocessing

```python
data['Item_List'] = data['Items'].str.split(',')
data ['Item_List']
```

```python
from mlxtend.preprocessing import TransactionEncoder
```

```python
# Assuming 'data' is already loaded and contains the transactions split into lists

te = TransactionEncoder()
te_ary = te.fit(data['Item_List']).transform(data['Item_List'])
df = pd.DataFrame(te_ary, columns=te.columns_)
df
```

```python
df.replace(to_replace = True, value = 1, inplace=True)
df.replace(to_replace = False, value = 0, inplace=True)
```

```python
df
```

# Apriori Algorithm

```python
from mlxtend.frequent_patterns import apriori, association_rules

frequent_itemsets = apriori(df, min_support=0.005, max_len=3, use_colnames = True)  #support tells us no. of times given transaction occur in dataset
frequent_itemsets
```

## max_len ?
- The `max_len` parameter in the above code specifies the maximum number of items that can be present in a frequent itemset.
- In other words, it controls the size of the itemsets that are considered for analysis.

- `max_len` parameter is set to `3`. This means that the `apriori` function will only consider itemsets that contain up to 3 items.
- For example, the itemset `{Milk, Bread, Eggs}` would be considered, but the itemset `{Milk, Bread, Eggs, Butter}` would not.

- By adjusting the `max_len` parameter, you can control the complexity of the analysis and the size of the output.
- A larger `max_len` will result in more complex itemsets being considered, but it will also increase the computational cost of the analysis.


```python
rules = association_rules(frequent_itemsets, metric='lift',min_threshold=1) # min_threshold= threshold should be atleast 0.7
rules.sort_values('lift',ascending=False).head(10) # select associations with high confidence, if lift ratio's are similar
```


## min_threshold ?
- The `min_threshold` parameter in the `association_rules` function specifies the minimum confidence level for an association rule to be considered significant.

- In other words, it controls the strength of the relationship between the items in the rule.

- A higher `min_threshold` will result in fewer rules being generated, but the rules that are generated will have a stronger relationship between the items.

- For example, if you set `min_threshold` to 0.8, then only rules where the confidence is at least 80% will be considered significant.

- The default value for `min_threshold` is 0.5, which means that only rules where the confidence is at least 50% will be considered significant.

- In the code, the `min_threshold` is set to 1. This means that only rules where the confidence is 100% will be considered significant.

- This is a very high threshold, and it will likely result in very few rules being generated.

# **Interview Questions:**

**1. What is lift and why is it important in Association rules?**
* Sometimes, association rules gives high confidence due to intrinsic connection btw them
* To avoid this we use another measure called `benchmark confidence`
* `Lift Ratio ` = `confidence` / `benchmark confidence`
* Higer the `Lift Ratio` better is the association

<br>

**2. What is support and Confidence. How do you calculate them?**
* `support` : support tells us no. of times given transaction occur in dataset
* `support` = `Transactions with both Antecedent & Consequent` / `Total no. of Transactions`
* With `min_support` = x %, we can eliminate some associations

*  `confidence` : Likelihood of having both Antecedent & Consequent in the dataset, when Antecedent is already present
* `confidence` = `No. of Transaction with both Antecedent & Consequent` / `No. of Items with Antecedent`

**3.	What are some limitations or challenges of Association rules mining?**

* **Data quality:**
  - The quality of the data used for association rule mining can have a significant impact on the results.
  - If the data is noisy, incomplete, or inaccurate, the resulting association rules may be unreliable.
* **Computational complexity:**
  - Association rule mining can be computationally expensive, especially for large datasets.
  - This is because the number of possible itemsets grows exponentially with the number of items in the dataset.
* **False positives:**
  - Association rule mining can generate a large number of false positives, which are rules that appear to be significant but are actually not.
  - This is because the rules are based on a statistical analysis of the data, and there is always a chance that some of the rules will be false positives.
* **Overfitting:**
  - Association rule mining can overfit the data, which means that the rules are too specific to the particular dataset and do not generalize well to other datasets.
  - This can be a problem if the dataset is small or if it is not representative of the population of interest.

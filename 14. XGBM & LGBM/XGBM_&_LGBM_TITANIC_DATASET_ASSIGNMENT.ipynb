# LGBM

```python
import lightgbm as lgb
d_train = lgb.Dataset(x_train, label=y_train)
```

```python
params = {}
params['learning_rate'] = 0.003
params['boosting_type'] = 'gbdt'
params['objective'] = 'binary'
params['metric'] = 'binary_logloss'
params['sub_feature'] = 0.5
params['num_leaves'] = 10
params['min_data'] = 50
params['max_depth'] = 10
```

# what is metric , sub_feature, min_data above

**Metric:**
- The metric parameter in the LightGBM model specifies the evaluation metric used to measure the model's performance. - In this case, the metric is set to 'binary_logloss', which is a common metric for binary classification problems.

**Sub_feature:**
- The sub_feature parameter in the LightGBM model controls the proportion of features used in each tree split.
- A value of 0.5 means that 50% of the features are used at each split.
- This parameter can help to prevent overfitting and improve modelgeneralizability.

**Min_data:**
- The min_data parameter in the LightGBM model specifies the minimum number of data points required in a leaf node.    
- This parameter can help to prevent overfitting by ensuring that each leaf node has a sufficient amount of data to make a reliable prediction.

```python
clf = lgb.train(params, d_train, 500)
```

```python
# predictions
y_pred = clf.predict(x_train)
```

```python
predictions = [round(value) for value in y_pred]
```

```python
predictions
```

```python
accuracy = accuracy_score(y_train, predictions)
print('Accuracy %.2f'% (accuracy * 100))
```

# Hyperparameter tuning for XG Boost

```python
from sklearn.model_selection import GridSearchCV

# Define Hyperparameter Grid
param_grid = {
    'max_depth': [3, 4, 5],
    'learning_rate': [0.1, 0.01, 0.001],
    'n_estimators': [50, 100, 200],
    'subsample': [0.5, 0.7, 1.0],
    'colsample_bytree': [0.5, 0.7, 1.0]
}

# Initialize XGBoost Classifier
xgb_model = XGBClassifier()

# Initialize Grid Search with cross validation
grid_search = GridSearchCV(estimator=xgb_model, param_grid=param_grid, cv=5, scoring='accuracy')

# Perform Grid Search
grid_search.fit(x_train, y_train)

# Get the best parameters and best score
best_params = grid_search.best_params_
best_score = grid_search.best_score_

# Train final model with best parameters
final_model = XGBClassifier(**best_params)
final_model.fit(x_train, y_train)
```

```python
print(best_params)
print(best_score)
```

# Now rebuild the model using these parameters

```python
params = {
    'max_depth': 4,
    'learning_rate': 0.1,
    'n_estimators': 200,
    'subsample': 0.5,
    'colsample_bytree': 0.5
}

model = XGBClassifier(**params)
clf_xgb = model.fit(x_train, y_train)
y_pred_train_xgb = clf_xgb.predict(x_train)
acc_xgb = round(clf_xgb.score(x_train, y_train)*100, 2)
print(acc_xgb)
```

```python
y_train.value_counts()
```

```python
from sklearn.metrics import precision_score, recall_score, f1_score

precision = precision_score(y_train, y_pred_train_xgb)
recall = recall_score(y_train, y_pred_train_xgb)
f1 = f1_score(y_train, y_pred_train_xgb)

print("Precision:", precision)
print("Recall:", recall)
print("F1 Score:", f1)
```
